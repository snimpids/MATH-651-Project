```python
from edge_prog_1d import *
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='serif', serif='Times')
plt.rc('')
```

```python
g1, t = gaus1(256,1,1)
w = np.linspace(-5,5,256)
g1_fft = 1j*w*np.exp(-1*1**2*w**2/2)
g1_fft = np.abs(g1_fft)
```

```python
fig, [ax1, ax2] = plt.subplots(ncols=2, figsize=[6,2.5])
ax1.plot(t,-g1,'k')
ax1.set_xlabel(r'$t$')
ax1.set_title(r'$\psi(t)$')
ax1.grid(True)

ax2.plot(w,g1_fft,'k')
ax2.set_xlabel(r'$w$')
ax2.set_title(r'$\hat\psi(w)$')
ax2.grid(True)

plt.tight_layout()

fig.savefig('/Users/lida/Desktop/identifying-buildings-satellite/report_image/edge1_gaus1.pdf', dpi=300)
```

```python
N = 256; t = np.linspace(0,N-1,N);
sig2 = np.zeros(N)
sig2[20:40] = 1; sig2[40:50] = 2
sig2[80:130] = -2/625 * (t[80:130]-105)**2 + 2;
sig2[150] = 2;
sig2[180:225] = 2/45**2 * (t[180:225]-225)**2
```

```python
scale = np.linspace(-6, -2, 16);
scale = 2**scale
```

```python
C2 = cwt_1d(sig2, scale)
MM2 = wtmm_1d(C2)
```

```python
t = np.linspace(0,1,N)
p6, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=[6,6])

ax0.plot(t,sig2,'k')
ax0.autoscale(enable=True, axis='x', tight=True)
ax0.set_title('(a) Test Signal')
ax0.set_ylim([-1,3])
ax0.grid(True)

ax1.pcolor(t, scale, C2, cmap='gray', linewidth=0,rasterized=True)
ax1.set_yscale("log", nonposx='clip', basey=2)
ax1.invert_yaxis()
ax1.set_title('(b) Continuous Wavelet Transform')
ax1.set_ylabel('scale')
ax1.grid(False)

ax2.pcolor(t, scale, MM2, cmap='gray', linewidth=0 ,rasterized=True)
ax2.set_yscale("log", nonposx='clip', basey=2)
ax2.invert_yaxis()
ax2.set_title('(c) Wavelet Transform Modulus Maximum')
ax2.set_xlabel('time (s)')
ax2.set_ylabel('scale')
ax2.grid(False)

plt.tight_layout()
p6.savefig('/Users/lida/Desktop/identifying-buildings-satellite/report_image/edge1_cwt1.pdf', dpi=300)
```
