# Show Edge Detection Result
This is a notebook to show edge detection result, and also make graph for the final report.

```python
from edge_prog_2d import *
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## Read the Satellite Image

```python
I = plt.imread('./png_satellite_image/6100_0_3_gray.png')
fig = plt.subplots(figsize=[12,8])
plt.imshow(I, cmap='gray')
```

```python
N = 256
x1 = 300 
y1 = 1140
x2 = 2200
y2 = 2230
x3 = 1800
y3 = 350

I1 = I[x1:x1+N-1, y1:y1+N-1]
I1 = I1[:,:,1]
I2 = I[x2:x2+N-1, y2:y2+N-1]
I2 = I2[:,:,1]
I3 = I[x3:x3+N-1, y3:y3+N-1]
I3 = I3[:,:,1]

plt.imsave('./edge_data/image1.png', I1, cmap='gray')
plt.imsave('./edge_data/image2.png', I2, cmap='gray')
plt.imsave('./edge_data/image3.png', I3, cmap='gray')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=[12,6]);
ax1.imshow(I1, cmap='gray')
ax1.set_title('Image 1')
ax1.set_xticks([]);
ax1.set_yticks([]);

ax2.imshow(I2, cmap='gray')
ax2.set_title('Image 2')
ax2.set_xticks([]);
ax2.set_yticks([]);

ax3.imshow(I3, cmap='gray')
ax3.set_title('Image 3')
ax3.set_xticks([]);
ax3.set_yticks([]);
```

## Satellite Image Edge Detection
Work flow:
1. Read the image data
2. Make a scale sequence
3. Do CWT on each scale, get E
4. Calculate Lipschite Exponent

### Image 1

```python
scale = np.linspace(-6, -3, 16);
scale = 2**scale

E1 = cal_multi_edge(I1, scale)

a, ind, val, s = LE_2d(E1, scale, 0.1)

th = np.linspace(-0.6,-0.35,6)
Lip1 = np.ndarray([E1.shape[0], E1.shape[1], 6])
E_one1 = np.ndarray([E1.shape[0], E1.shape[1], 6])
E_val1 = np.ndarray([E1.shape[0], E1.shape[1], 6])
for i in range(6):
    Lip1[:,:,i], E_one1[:,:,i], E_val1[:,:,i] = ex_edge_by_Lip(E1, a, ind, th[i])
    
    L_name = './edge_data/image1_lip_' + str(i) + '_' + str(th[i]) + '.png'
    plt.imsave(L_name, Lip1[:,:,i], cmap='gray')
    Eone_name = './edge_data/image1_Eone_' + str(i) + '_' + str(th[i]) + '.png'
    plt.imsave(Eone_name, E_one1[:,:,i], cmap='gray')
    Eval_name = './edge_data/image1_Eval_' + str(i) + '_' + str(th[i]) + '.png'
    plt.imsave(Eval_name, E_val1[:,:,i], cmap='gray')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=[14,7])
for ax,i in zip(axes.flat, range(6)):
    ax.imshow(Lip1[:,:,i], cmap='gray')
    title = 'Image 1, LE <'
    title += str(th[i])
    ax.set_title(title)
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=[14,7])
for ax,i in zip(axes.flat, range(6)):
    ax.imshow(E_val1[:,:,i], cmap='gray')
    title = 'Image 1, LE <'
    title += str(th[i])
    ax.set_title(title)
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=[14,7])
for ax,i in zip(axes.flat, range(6)):
    ax.imshow(E_one1[:,:,i], cmap='gray')
    title = 'Image 1, LE <'
    title += str(th[i])
    ax.set_title(title)
    ax.set_xticks([])
    ax.set_yticks([])
```
